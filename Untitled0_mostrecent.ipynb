```
df = pd.read_csv('train.csv')
tagger = MeCab.Tagger()
sw=list(stop_df['stop'])
# my_stop=["の", "が", "に", "た", "で", "と", "を", "は", "て", "も", "へ",\
#          "、", "。", "「", "」", "！", "？", "・", "『", "』", "=", "-", \
#          "(", ")"'「','」','【','】','「','※',':','」','［','］',':','（','）']
# sw=sw+my_stop

def normalize_number(text):
    """
    pattern = r'\d+'
    replacer = re.compile(pattern)
    result = replacer.sub('0', text)
    """
    # 連続した数字を0で置換
    replaced_text = re.sub(r'\d+', '0', text)
    return replaced_text
  
  

# 入力されたテキストを単語単位に分割して返却する関数
def parse_text(text):
  
    words = []
    for morph in tagger.parse(text).rstrip().split("\n")[:-1]:
      word=morph.split("\t")[0]
      if not word in sw:
         words.append(word)
    return " ".join(words)

  
  
# df["title"] = df.title.apply(normalize_number)
# df["content"] = df.content.apply(normalize_number)

df["title_words"] = df.title.apply(parse_text)
df["content_words"] = df.content.apply(parse_text)

print('done')
few_text_row=150>df['content'].apply(lambda x:len(x))
df.loc[few_text_row,"content_words"]+=' '+df.loc[few_text_row,"title_words"]





# stop_words = set([
#     "の", "が", "に", "た", "で", "と", "を", "は", "て", "も", "へ",
#     "、", "。", "「", "」", "！", "？", "・", "『", "』", "=", "-", "(", ")"
# ])

# df.groupby("media_id")["title_words"].apply(
#     lambda x: Counter([w for w in " ".join(x).split(" ") if w not in stop_words]).most_common(n=10)
# )[:25]
```
